```
def addIndex(df, month):
  order = []
  i = 1
  for value in df["Page"]:
    order.append(i)
    i += 1
  df[month] = order
  return df
```

```
data_sep = addIndex(data_sep, "Sep")
data_sep.drop('Pageviews', 1, inplace=True)
```

```
data_oct = addIndex(data_oct, "Oct")
data_oct.drop('Pageviews', 1, inplace=True)
```

```
data_nov = addIndex(data_nov, "Nov")
data_nov.drop('Pageviews', 1, inplace=True)
```

```
data_dec = addIndex(data_dec, "Dec")
data_dec.drop('Pageviews', 1, inplace=True)
```

```
data_jan = addIndex(data_jan, "Jan")
data_jan.drop('Pageviews', 1, inplace=True)
```

```
data_feb = addIndex(data_feb, "Feb")
data_feb.drop('Pageviews', 1, inplace=True)
```

```
data_mar = addIndex(data_mar, "Mar")
data_mar.drop('Pageviews', 1, inplace=True)
```

```
data_apr = addIndex(data_apr, "Apr")
data_apr.drop('Pageviews', 1, inplace=True)
```

```
data_may = addIndex(data_may, "May")
data_may.drop('Pageviews', 1, inplace=True)
```

```
data_jun = addIndex(data_jun, "Jun")
data_jun.drop('Pageviews', 1, inplace=True)
```

```
data_jul = addIndex(data_jul, "Jul")
data_jul.drop('Pageviews', 1, inplace=True)
```

```
data_aug = addIndex(data_aug, "Aug")
data_aug.drop('Pageviews', 1, inplace=True)
```

```
data_combined = pd.merge(data_sep, data_oct, on='Page')
data_combined = pd.merge(data_combined, data_nov, on='Page')
data_combined = pd.merge(data_combined, data_dec, on='Page')
data_combined = pd.merge(data_combined, data_jan, on='Page')
data_combined = pd.merge(data_combined, data_feb, on='Page')
data_combined = pd.merge(data_combined, data_mar, on='Page')
data_combined = pd.merge(data_combined, data_apr, on='Page')
data_combined = pd.merge(data_combined, data_may, on='Page')
data_combined = pd.merge(data_combined, data_jun, on='Page')
data_combined = pd.merge(data_combined, data_jul, on='Page')
data_combined = pd.merge(data_combined, data_aug, on="Page")
#data_combined.head()
```

```
maxValues = data_combined.max(axis=1)
minValues = data_combined.min(axis=1)
diffValues = maxValues - minValues
data_combined["Diff"] = diffValues

#print(data_combined)
```

```
alt.Chart(data_combined, height=1000).mark_point().encode(
    alt.X('Page', sort='y', axis=alt.Axis(labelAngle=-45)),
    y='Diff',
    tooltip=['Page', 'Sep', 'Oct', 'Nov', 'Dec', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Diff'],
    color=alt.Color('Page', legend=None)
    )
```
